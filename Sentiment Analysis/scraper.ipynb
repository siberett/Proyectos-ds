```python
import numpy as np
import pandas as pd
import snscrape.modules.twitter as sntwitter
```

```python
query = '(@Uber) lang:en until:2017-03-20 since:2017-01-20'

tweets = []
for tweet in sntwitter.TwitterSearchScraper(query).get_items():
    
    tweets.append([tweet.id, tweet.date, tweet.user.username, tweet.rawContent, tweet.user.location,
                      tweet.retweetCount, tweet.likeCount])
        
uberdf = pd.DataFrame(tweets, columns=['id', 'date', 'username', 'content', 'location', 'retweets', 'likes'])
```

```python
uberdf
```

```python
uberdf.to_csv('uberdf.csv')
```
